```python
table["percent"]  = table.apply(
lambda row: row[True] / (row[True] + row[False]) * 100, axis=1 
)
table
```

```python
df["date"] = pd.to_datetime(df["date"])
df["weekday"] = df["date"].dt.day_name()
orders = ["Monday","Thusday","Wendsay","Friday","Saturday","Sunday"]
df["weekday"] = pd.Categorical(df["weekday"], categories=orders, ordered=True)
counts = df["weekday"].value_counts().sort_index()
counts
```

```python
import requests

url = "https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
response.raise_for_status()  

state_and_territory_abbreviations = pd.read_html(response.text)
state_and_territory_abbreviations[1]

```

```python
url = "https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
response.raise_for_status()

states_by_population = pd.read_html(response.text)
states_by_population[0]
```

```python
import pandas as pd

df_incidents = df
df_pop = states_by_population[0]
df_abbrev = state_and_territory_abbreviations[1]

df_abbrev = df_abbrev[['Name', 'USPS (& ANSI)']]
df_abbrev.columns = ['state_full', 'state_code']
df_abbrev = df_abbrev[df_abbrev['state_code'].str.len() == 2]

df_pop = df_pop[['State', 'Census population, April 1, 2020 [1][2]']]
df_pop.columns = ['state_full', 'population']

df_pop['population'] = (
    df_pop['population']
    .astype(str)
    .str.replace(',', '')
    .astype(int)
)

df_states = pd.merge(df_pop, df_abbrev, on='state_full', how='inner')

df_counts = df_incidents.groupby('state').size().reset_index(name='incidents')

df_final = pd.merge(df_states, df_counts, left_on='state_code', right_on='state', how='left')

df_final['incidents'] = df_final['incidents'].fillna(0)
df_final = df_final.drop(columns=['state'])

df_final['incidents_per_1000'] = df_final['incidents'] / df_final['population'] * 1000

df_final.sort_values('incidents_per_1000', ascending=False)



```
